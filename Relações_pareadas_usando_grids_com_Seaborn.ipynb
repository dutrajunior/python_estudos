```python
grid = sns.PairGrid(data=df, hue="gender")
grid = grid.map_upper(plt.scatter).add_legend()
grid = grid.map_diag(plt.hist)
grid = grid.map_lower(sns.kdeplot)
```

```python
facet = sns.FacetGrid(df, col="test preparation course")
#facet.map(sns.distplot.kde(True), "writing score")
facet.map_dataframe(sns.histplot, x="writing score", kde=True, stat="density")
```

```python
facet = sns.FacetGrid(df, col="test preparation course", hue="lunch")
facet.map_dataframe(sns.histplot, x="writing score", kde=True, stat="density").add_legend()
```

```python
facet = sns.FacetGrid(df, col="test preparation course", row="race/ethnicity", hue="lunch", margin_titles=True)
facet.map_dataframe(sns.histplot, x="writing score", kde=True, stat="density").add_legend()
```

```python
facet = sns.FacetGrid(df, col="test preparation course", hue="lunch")
facet.map(plt.scatter, "writing score", "reading score").add_legend()
```

```python
facet = sns.FacetGrid(df, col="test preparation course", hue="lunch")
facet.map(plt.scatter, "writing score", "math score").add_legend()
```

```python
df.describe()
```

```python
grid = sns.FacetGrid(df, row='smoker', col='time')
grid.map(sns.scatterplot, 'size', 'total_bill')
```

```python
grid = sns.FacetGrid(df, row='smoker', col='time', margin_titles=True)
grid.map(sns.scatterplot, 'size', 'total_bill')
```

```python
grid = sns.FacetGrid(df, col='day')
grid.map(sns.barplot, 'sex', 'total_bill')
```

```python
dias_ordenados = ['Thur', 'Fri', 'Sat', 'Sun']
grid = sns.FacetGrid(df, col='day', col_order=dias_ordenados)
grid.map(sns.barplot, 'sex', 'total_bill')
```

```python
grid = sns.FacetGrid(df, col='sex', hue='day')
grid.map(sns.scatterplot, 'tip', 'total_bill').add_legend()
```

```python
cores = dict(Thur = 'red', Fri='orange', Sat='purple', Sun='green')
grid = sns.FacetGrid(df, col='sex', hue='day', palette=cores)
grid.map(sns.scatterplot, 'tip', 'total_bill').add_legend()
```

```python
def hexbin(x, y, color, **kwargs):
  cmap = sns.light_palette(color, as_cmap=True)
  plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

grid = sns.FacetGrid(df, col='time', hue='time')
grid.map(hexbin, 'total_bill', 'tip', extent=[0, 50, 0, 10])
```

```python
#criar coluna wine_type
red_wine['wine_type'] = 'red'
white_wine['wine_type'] = 'white'

#unificar as tabelas
wines = pd.concat([red_wine, white_wine])
wines
```

```python
df_high = wines.loc[wines['quality']> 7]
df_high
```

```python
cores = dict(red = 'red', white='green')

grid = sns.PairGrid(data=df_high[['wine_type', 'density','residual sugar', 'pH','sulphates', 'alcohol']], hue='wine_type', palette=cores)
grid = grid.map_upper(plt.scatter)
grid = grid.map_diag(plt.hist)
grid = grid.map_lower(sns.kdeplot).add_legend()
```
