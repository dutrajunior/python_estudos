```python
sns.histplot(dataframe['Age'].dropna())
plt.xlabel('Age')
plt.ylabel('Quantidade')
plt.show()
```

```python
sns.kdeplot(dataframe['Age'].dropna())
plt.xlabel('Age')
plt.ylabel('Densidade')
plt.show()
```

```python
sns.histplot(dataframe['Age'].dropna(), kde=True, stat="density")
plt.xlabel('Age')
plt.ylabel('Densidade')
plt.show()
```

```python
sns.histplot(dataframe['Age'].dropna(), kde=True)
plt.xlabel('Age')
plt.ylabel('Quantidade')
plt.show()
```

```python
dataframe = dataframe.dropna(subset=['Embarked'])
dataframe = dataframe.dropna(subset=['Fare'])

Embarkeds = ['S', 'C', 'Q']

for Embarked in Embarkeds:
  subset = dataframe[dataframe['Embarked'] == Embarked]

  sns.kdeplot(data=subset['Fare'], label = Embarked)


plt.legend(prop={'size':6}, title = 'Embarked')
plt.title('Grafico de Densidade')
plt.xlabel('Avaliação')
plt.ylabel('Densidade')
plt.show()
```

```python
sns.jointplot(x='Age', y='Fare', data=dataframe)
plt.show()
```

```python
sns.jointplot(x='Age', y='Fare', data=dataframe, kind='kde')
plt.show()
```
