# Importando as bibliotecas b√°sicas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/dutrajunior/python_estudos/main/loan_data.csv')
dados
```

# Aplicando o Label Encoder em Loop nas Colunas Object

Itemos aplicar o Label Encoder porque o SVM ir√° se basear em valores num√©ricos para definir o melhor plano, ent√£o ele s√≥ sabe trabalhar com vari√°veis num√©ricas. Iremos transformar a coluna Purpose em num√©rica para continuar

```python
dados.dtypes
```

```python
#Tranformar o resultado do dtypes em um dataframe com o uso do reset_index()
colunas = dados.dtypes.reset_index()
colunas
```

```python
colunas.columns
```

```python
#buscar somente as colunas tipo object
colunas[colunas[0] == 'object']['index']
```

```python
#colocar numa lista as colunas object
categ_cols = colunas[colunas[0] == 'object']['index'].to_list()
categ_cols
```

```python
# Fazendo o Label Encoder em loop para todas as colunas object
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in categ_cols:
    le.fit(dados[i])
    dados[str(i) + '_encoded'] = le.transform(dados[i])
```

```python
dados
```

```python
#remover a coluna object do dataframe orginal
dados = dados.drop('purpose', axis = 1)
dados
```

# Implementando o modelo

Vari√°vel que queremos prever √© a Not Fully Paid (ou seja, se a pessoa n√£o pagou totalmente o impr√©stimo)


## Separando a base em X, vari√°veis de entrada e Y, vari√°vel que queremos prever (Not Fully Paid)

```python
x = dados.drop('not.fully.paid', axis = 1) # base original removendo a coluna "Not Fully Paid"
y = dados['not.fully.paid'] # Dados somente da coluna "Not Fully Paid"
```

##  Quebrando o dataset em treino e teste - Iremos usar 70% para treino e 30% para teste

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Carregando o modelo

```python
from sklearn.svm import SVC

# Selecionando qual modelo eu quero treinar, o kernel iremos manter como linear,
# mas importante pesquisar sobre os outros tipos de kernel para o SVM (sigmoide, radial e polynomial)

svm = SVC(kernel='linear')
svm.fit(x_train, y_train)

# Fazendo a previs√£o

y_pred = svm.predict(x_test)
```

## Avaliando o Resultado

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```
