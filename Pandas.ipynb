## JunÃ§Ã£o de dados

**append**

*Consta em depreciaÃ§Ã£o*

Anexa um conjunto de dados ao final do dataframe selecionado, tem a opÃ§Ã£o de manter (ignore_index) os indices originais ou continuar o indice da tabela principal

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
df2
```

```python
result = df1.append(df2, ignore_index=True)
result
```

### Pandas concatenar

Juntar conjuntos registros de conjuntos de dados

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
df2
```

Concatena dois dataframes (axis=0)

tem a opÃ§Ã£o de manter (ignore_index) os indices originais ou continuar o indice da tabela inicial

```python
result = pd.concat([df1, df2],axis=0, ignore_index=True)
result
```

Juntar colunas de conjuntos de dados (axis=1)


```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],})
df1
```

```python
df2 = pd.DataFrame({'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
df2
```

```python
result = pd.concat([df1, df2],axis=1)
result
```

**join**

```python
df_left = pd.DataFrame({"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"])
df_left
```

```python
df_right = pd.DataFrame({"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"])
df_right
```

```python
result = df_left.join(df_right)
result
```

```python
result = df_left.join(df_right, how="outer")
result
```

```python
result = df_left.join(df_right, how="inner")
result
```

```python
df_left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"],
    }
)


df_right = pd.DataFrame({"C": ["C0", "C1"], "D": ["D0", "D1"]}, index=["K0", "K1"])

result = df_left.join(df_right, on="key")
result
```
