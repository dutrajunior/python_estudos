```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips.shape
```

```python
tips.describe()
```

```python
## GrÃ¡fico de Barras
ax0 = sns.barplot(x="day", y="total_bill", data=tips)
```

```python
## GrÃ¡fico de DispersÃ£o
ax1 = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

# Uso de cores

```python
ax2 = sns.barplot(x="day", y="total_bill", data=tips, color="coral")
```

```python
ax3 = sns.scatterplot(x="total_bill", y="tip", data=tips, color = 'greenyellow')
```

```python
ax4 = sns.barplot(x="day", y="total_bill", data=tips, palette="Blues_d")
```

```python
cores = dict(Thur = "#3baee5", Fri="#5488a5", Sat="#e18b2b", Sun="#e3a839")
ax5 = sns.barplot(x="day", y="total_bill", data=tips, palette=cores)
```

```python
cores = dict(Lunch = "gold", Dinner="midnightblue")
ax6 = sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", palette=cores)
```

# ModificaÃ§Ã£o de formas

```python
ax7 = sns.scatterplot(x="total_bill", y="tip", data=tips, marker='+')
```

```python
marcadores = dict(Lunch="D", Dinner=",")
ax8 = sns.scatterplot(x="total_bill", y="tip", data=tips, style="time", markers=marcadores)
```

# ModificaÃ§Ã£o de tamanhos

```python
ax9 = sns.scatterplot(x="total_bill", y="tip", data=tips, size="size")
```

# AlteraÃ§Ã£o de escalas

```python
ax10 = sns.barplot(x="day", y="total_bill", data=tips)
plt.ylim(15, 22)
```

```python
ax11 = sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.ylim(2, 4)
plt.xlim(10, 30)
```
