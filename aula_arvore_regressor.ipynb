# Importando as bibliotecas b√°sicas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importando o dataset

```python
dados = pd.read_csv(r'https://raw.githubusercontent.com/dutrajunior/python_estudos/main/house_prices.csv')
```

# Olhando para os dados

```python
dados
```

```python
dados.describe()
```

Iremos preencher os dados nulos para cara coluna devido a termos uma quantidade significativa de nulos na base

```python
colunas = dados.dtypes.reset_index()
colunas
```

```python
colunas = dados.dtypes.reset_index()

cont_cols = colunas[colunas[0] != 'object']['index'].to_list()

# Preenchendo as vari√°veis com dados nulos com a m√©dia

for i in cont_cols:
    dados[i] = dados[i].fillna(dados[i].mean())
```

```python
dados.describe()
```

Iremos prever o resultado da coluna SalePrice, ou seja, temos uma s√©rie de caracter√≠sticas de casas e queremos saber o pre√ßo final.

Para isso iremos aplicar uma modalidade da √°rvore de decis√£o que √© o Regressor (para regress√µes)

# Aplicando o Label Encoder

```python
colunas = dados.dtypes.reset_index()

categ_cols = colunas[colunas[0] == 'object']['index'].to_list()

# Fazendo o Label Encoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in categ_cols:
    le.fit(dados[i])
    dados[str(i) + '_encoded'] = le.transform(dados[i])
    dados = dados.drop(i, axis = 1)
```

```python
dados
```

# Quebrando o dataset para modelagem

```python
from sklearn.model_selection import train_test_split

x = dados.drop('SalePrice', axis = 1)
y = dados['SalePrice']

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

# Carregando o modelo

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score


modelo = DecisionTreeRegressor(max_depth=10, max_leaf_nodes=20)

modelo.fit(x_train, y_train)
y_pred_train = modelo.predict(x_train)
y_pred_test = modelo.predict(x_test)

print(r2_score(y_train,y_pred_train))
```

Verificando o resultado do teste

```python
print(r2_score(y_test,y_pred_test))
```

Fazendo uma tentativa de variar os par√¢metros da √°rvore

```python
depth = [10, 50, 100, 500]
leaf = [20, 50, 100, 500]

for i,j in zip(depth, leaf):

    print('Depth = ', i)
    print('Leaf = ', j)
    modelo = DecisionTreeRegressor(max_depth=i, max_leaf_nodes=j, random_state=42)
    modelo.fit(x_train, y_train)
    y_pred_train = modelo.predict(x_train)
    y_pred_test = modelo.predict(x_test)

    print('Score de Treino √© :',r2_score(y_train,y_pred_train))
    print('Score de Teste √© :',r2_score(y_test,y_pred_test))
```

O Modelo performou melhor por volta de 100 nos par√¢metros testados. √â poss√≠vel notar um Overfitting para par√¢metros com valores muito elevados
