```python
BasePetal = BaseIris[['petal_width','petal_length', 'target']]
BasePetal
```

```python
BasePetalK = BaseIris[['petal_width','petal_length']]
BasePetalK
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(BasePetalK)
```

```python
BasePetalK['target'] = kmeans.labels_
BasePetalK
```

```python
def autoLabelEncoder(database):
  #colocar numa lista as colunas object
  AllColumns = database.dtypes.reset_index()
  ObjectColumns = AllColumns[AllColumns[0] == 'object']['index'].to_list()

  # Fazendo o Label Encoder em loop para todas as colunas object
  from sklearn.preprocessing import LabelEncoder

  le = LabelEncoder()

  for i in ObjectColumns:
      le.fit(database[i])
      #database[str(i) + '_encd'] = le.transform(database[i])
      database[i] = le.transform(database[i])
      database = database.drop(str(i), axis = 1)

  return database
```

```python
BasePetalEncd = autoLabelEncoder(BasePetal)
```

```python
#alteraÃ§Ã£o dos rotulos dos clusters para ficarem similares
BasePetalK.loc[BasePetalK['target'] == 1, 'target'] = 4
BasePetalK.loc[BasePetalK['target'] == 2, 'target'] = 1
BasePetalK.loc[BasePetalK['target'] == 4, 'target'] = 2
```

```python
sb.scatterplot(x="petal_width", y="petal_length", data=BasePetalK, hue='target')
pl.show()
```

```python
sb.scatterplot(x="petal_width", y="petal_length", data=BasePetalEncd, hue='target')
pl.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(BasePetalEncd['target'], BasePetalK['target'])
```
