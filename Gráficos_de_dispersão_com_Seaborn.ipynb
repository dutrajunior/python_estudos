```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
gorjeta = sn.load_dataset('tips')
gorjeta
```

```python
gorjeta.head()
```

```python
gorjeta.tail()
```

```python
gorjeta.shape
```

```python
gorjeta.dtypes
```

```python
gorjeta.describe()
```

```python
sn.relplot(data=gorjeta, x="total_bill", y='tip', hue='sex')
```

```python
sn.scatterplot(data=gorjeta, x="total_bill", y='tip', hue='day', style='time')
```

```python
sn.catplot(data=gorjeta, x="day", y='total_bill', kind='swarm')
```

```python
sn.lmplot(data=gorjeta, y='tip', hue='smoker', col="time", palette='Set1', scatter_kws={'s':90, 'alpha':0.5}, line_kws={'lw':2})
```

```python
ax = sn.scatterplot(x='price', y='horsepower', data=df)
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sn.lmplot(x='price', y='horsepower', data=df, fit_reg=False, hue='fueltype')
```

```python
sn.lmplot(x='price', y='horsepower', data=df, fit_reg=True, hue='fueltype')
```

```python
sn.catplot(x='fuelsystem', y='price', data=df)
```

```python
sn.catplot(x='fuelsystem', y='price', data=df, jitter = False)
```

```python
sn.catplot(x='fuelsystem', y='price', data=df, hue = 'fueltype')
```

```python
sn.catplot(x='fuelsystem', y='price', hue = 'fueltype', data=df, jitter = False)
```

```python
sn.catplot(x='fuelsystem', y='price', hue = 'fueltype', data=df, kind = 'box')
```

```python
sn.catplot(x='fuelsystem', y='price', hue = 'fueltype', data=df, kind = 'boxen')
```

```python
sn.catplot(x='fuelsystem', y='price', hue = 'fueltype', data=df, kind = 'boxen')
sn.stripplot(x='fuelsystem', y='price', color = 'green', data=df)
```
